```python
import numpy as np
import gym
from gym import spaces
import torch
import torch.nn as nn
import torch.optim as optim
from torch.distributions import Categorical, Normal
import matplotlib.pyplot as plt
import torch.nn.functional as F
import random
import pandas as pd
import os

SEED = 10
torch.manual_seed(SEED)
np.random.seed(SEED)
random.seed(SEED)

class TaskOffloadingEnv(gym.Env):
    def __init__(self, n_vehicles, d_V_values, bar_d_V_values, T_upload_C_values, T_download_C_values, F_C=20e9, max_steps=100):
        super(TaskOffloadingEnv, self).__init__()
        self.n = n_vehicles
        self.d_V_values = d_V_values
        self.bar_d_V_values = bar_d_V_values
        self.T_upload_C_values = T_upload_C_values
        self.T_download_C_values = T_download_C_values
        self.F_C = F_C
        self.max_steps = max_steps
        self.current_step = 0
        self.offloading_count = 0  # Initialize counter
        self.total_decisions = 0  # Total decisions made

        self.action_space = spaces.Tuple((
            spaces.MultiBinary(self.n),
            spaces.Box(low=2e9, high=4e9, shape=(self.n,), dtype=np.float32)
        ))

        self.observation_space = spaces.Box(low=0, high=np.inf, shape=(7 * self.n,), dtype=np.float32)

    def reset(self):
        self.current_step = 0
        self.offloading_count = 0  # Reset counter at the start of each episode
        self.total_decisions = 0  # Reset decision count

        self.d_V = np.random.randint(30040, 30041, self.n) * 8
        self.c_V = np.random.randint(130, 131, self.n) * 1e3 / 8 # 36000 sign + 94000 verify
        self.tau_V = np.random.randint(43, 44, self.n) * 0.000001 # 0.012 msec sign + 0.031 msec verify
        self.bar_d_V = np.random.randint(30040, 30041, self.n) * 8
        self.f_V = np.random.randint(1, 2, self.n) * 1e9 # hz
        self.T_upload_C = np.random.randint(100, 101, self.n) * 1e6 # 100 Mbps, 500Mbps, and 1000 Mbps
        self.T_download_C = np.random.randint(100, 101, self.n) * 1e6

        return np.concatenate([self.d_V, self.c_V, self.tau_V, self.bar_d_V, self.f_V, self.T_upload_C, self.T_download_C])

    def step(self, action):
        self.current_step += 1

        x_V, F_C_V = action
        x_V = x_V.reshape(-1)
        F_C_V = F_C_V.flatten()

        T_total = 0
        for i in range(self.n):
            self.total_decisions += 1
            if x_V[i] == 1:
                self.offloading_count += 1
                T_V_i = (self.d_V[i] / self.T_upload_C[i]) + \
                        ((self.d_V[i] * self.c_V[i]) / (F_C_V[i])) + \
                        (self.bar_d_V[i] / self.T_download_C[i]) + (2*6.67*0.000001) # 2km cloud tower distance
            else:
                T_V_i = (self.d_V[i] * self.c_V[i]) / self.f_V[i] # sec

            T_total += T_V_i

        x_V = torch.as_tensor(x_V, dtype=torch.float32)
        F_C_V = torch.as_tensor(F_C_V, dtype=torch.float32)

        if torch.sum(x_V * F_C_V) > self.F_C:
            penalty = -T_total * 0.1  # 10% of the total time as a penalty
            reward = -T_total + penalty
        else:
            reward = -T_total
    
        #if torch.sum(x_V * F_C_V) > self.F_C:
        #    reward = -T_total - 1e5 # penalty for exceeding capacity
        #else:
        #    reward = -T_total # sec

        next_state = np.concatenate([self.d_V, self.c_V, self.tau_V, self.bar_d_V, self.f_V, self.T_upload_C, self.T_download_C])
        done = self.current_step >= self.max_steps

        return next_state, reward, done, {}

class ActorCritic(nn.Module):
    def __init__(self, num_inputs, num_actions, hidden_size, std=0.2):
        super(ActorCritic, self).__init__()
        self.common = nn.Sequential(
            nn.Linear(num_inputs, hidden_size),
            nn.ReLU(),
            nn.Linear(hidden_size, hidden_size),
            nn.ReLU()
        )
        self.actor_fc = nn.Linear(hidden_size, num_actions)
        self.actor_fc2 = nn.Linear(hidden_size, num_actions)
        self.critic = nn.Linear(hidden_size, 1)
        self.log_std = nn.Parameter(torch.ones(1, num_actions) * std)
        
    def forward(self, x):
        x = self.common(x)

        probs = torch.sigmoid(self.actor_fc(x))
        means = self.actor_fc2(x)
        stds = self.log_std.expand_as(means)

        value = self.critic(x)

        return probs, means, stds, value

def compute_returns(next_value, rewards, masks, gamma):
    R = next_value
    returns = []
    for step in reversed(range(len(rewards))):
        R = rewards[step] + gamma * R * masks[step]
        returns.insert(0, R)
    return returns

def ppo_iter(mini_batch_size, states, actions, log_probs, returns, advantage):
    batch_size = states.size(0)
    for _ in range(batch_size // mini_batch_size):
        rand_ids = np.random.randint(0, batch_size, mini_batch_size)
        yield states[rand_ids, :], actions[rand_ids, :], log_probs[rand_ids, :], returns[rand_ids, :], advantage[rand_ids, :]

def ppo_update(model, optimizer, ppo_epochs, mini_batch_size, states, actions, log_probs, returns, advantages, entropy_coefficient, clip_param=0.2):
    for _ in range(ppo_epochs):
        for state, action, old_log_probs, return_, advantage in ppo_iter(mini_batch_size, states, actions, log_probs, returns, advantages):
            dist_probs, means, stds, value = model(state)
            action_continuous = action[:, n_vehicles:]

            dist_normal = Normal(means, stds)
            new_log_prob_continuous = dist_normal.log_prob(action_continuous).sum(1, keepdim=True)

            action_discrete = action[:, :n_vehicles]
            dist_categorical = Categorical(logits=dist_probs)

            action_discrete_i_list = []
            new_log_prob_discrete_list = []
            for i in range(n_vehicles):
                action_discrete_i = action_discrete[:, i]
                action_discrete_i_list.append(action_discrete_i.unsqueeze(-1))
                new_log_prob_discrete_i = dist_categorical.log_prob(action_discrete_i)
                new_log_prob_discrete_list.append(new_log_prob_discrete_i.unsqueeze(-1))

            new_action_discrete_i = torch.cat(action_discrete_i_list, dim=-1)
            new_log_prob_discrete_i = torch.cat(new_log_prob_discrete_list, dim=-1)

            new_log_probs = new_log_prob_discrete_i + new_log_prob_continuous

            ratio = (new_log_probs / (old_log_probs))
            surr1 = ratio * advantage
            surr2 = torch.clamp(ratio, 1.0 - clip_param, 1.0 + clip_param) * advantage

            actor_loss = -torch.min(surr1, surr2).mean()
            critic_loss = (return_ - value).pow(2).mean()

            optimizer.zero_grad()
            entropy = dist_categorical.entropy().mean() + dist_normal.entropy().mean()
            total_loss = actor_loss + 0.5 * critic_loss - entropy_coefficient * entropy
            
            total_loss.backward()

            max_norm = 0.5
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm)

            optimizer.step()

def evaluate_policy(env, model, n_vehicles, episodes=2):
    total_latency = 0.0
    for _ in range(episodes):
        state = env.reset()
        done = False
        while not done:
            state = torch.FloatTensor(state).unsqueeze(0)
            with torch.no_grad():
                dist_probs, means, stds, _ = model(state)
                action_cat = Categorical(dist_probs)
                action_discrete = action_cat.sample(sample_shape=torch.Size([n_vehicles]))
                action_discrete = action_discrete.reshape(1, n_vehicles)
                action_continuous = Normal(means, stds).sample()
                action = (action_discrete, action_continuous.numpy())
            next_state, reward, done, _ = env.step(action)
            total_latency += -reward
            state = next_state

    average_latency = total_latency / episodes
    return average_latency

def plot_latency(epochs_recorded, latencies, config_label):
    plt.figure(figsize=(10, 6))
    plt.plot(epochs_recorded, latencies, marker='o', linestyle='-', color='b', label='Latency')
    plt.xlabel('Iterations')
    plt.ylabel('Latency (msec)')
    plt.title(f'Latency vs Iterations for {config_label}')
    plt.grid(True)
    plt.legend()
    plt.show()


def run_experiment(n_vehicles, learning_rate, entropy_coefficient, gamma, d_V_values, bar_d_V_values, T_upload_C_values, T_download_C_values):
    env = TaskOffloadingEnv(n_vehicles=n_vehicles, d_V_values=d_V_values, bar_d_V_values=bar_d_V_values, T_upload_C_values=T_upload_C_values, T_download_C_values=T_download_C_values)
    num_inputs = env.observation_space.shape[0]
    num_actions = n_vehicles
    minimum_batch_size=64
    model = ActorCritic(num_inputs, num_actions, hidden_size=512)
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)

    max_frames = 3000
    frame_idx = 0
    latencies = []
    epochs_recorded = []

    state = env.reset()
    best_latency = float('inf')
    
    while frame_idx < max_frames:
        log_probs = []
        values = []
        states = []
        actions = []
        rewards = []
        masks = []
    
        for _ in range(100):
            state = torch.FloatTensor(state).unsqueeze(0)
            dist_probs, means, stds, value = model(state)
            dist_probs = dist_probs.squeeze(0)
            action_cat = Categorical(logits=dist_probs)
            action_discrete = action_cat.sample(sample_shape=torch.Size([n_vehicles]))
            action_discrete = action_discrete.reshape(1, n_vehicles)
            action_continuous = Normal(means, stds).sample()
            action_continuous_np = action_continuous.numpy()
            action = (action_discrete, action_continuous_np)
            next_state, reward, done, _ = env.step(action)
            log_prob = action_cat.log_prob(action_discrete) + Normal(means, stds).log_prob(action_continuous).sum(1, keepdim=True)
            
            states.append(state)
            actions.append(torch.cat([action_discrete, action_continuous], 1))
            rewards.append(reward)
            masks.append(1 - done)
            values.append(value)
            log_probs.append(log_prob)
    
            state = next_state
            frame_idx += 1
    
            if done:
                state = env.reset()
    
        next_state = torch.FloatTensor(next_state).unsqueeze(0)
        with torch.no_grad():
            _, _, _, next_value = model(next_state)
            
        returns = compute_returns(next_value, rewards, masks, gamma)
        returns = torch.cat(returns).detach()
        log_probs = torch.cat(log_probs).detach()
        values = torch.cat(values).detach()
        states = torch.cat(states)
        actions = torch.cat(actions)
        advantage = returns - values
        ppo_epoch = 4
        ppo_update(model, optimizer, ppo_epoch, minimum_batch_size, states, actions, log_probs, returns, advantage, entropy_coefficient)
    
        if frame_idx % 100 == 0:
            latency = evaluate_policy(env, model, n_vehicles)
            latencies.append(latency)
            epochs_recorded.append(frame_idx)
            avg_latency = np.mean(latencies[-100:])
            if avg_latency <= best_latency:
                best_latency = avg_latency
    
    # Calculate and print offloading percentage
    offloading_percentage = (env.offloading_count / env.total_decisions) * 100
    print(f'Offloading Percentage for {n_vehicles} vehicles: {offloading_percentage:.2f}%')

    config_label = f'n={n_vehicles}, lr={learning_rate}, hcoeff={entropy_coefficient}, gamma={gamma}, d_V={d_V_values}, bar_d_V={bar_d_V_values}, upload={T_upload_C_values}, download={T_download_C_values}'
    plot_latency(epochs_recorded, latencies, config_label)
    data = pd.DataFrame({
    'Iteration': epochs_recorded,
    'Latency at ' + config_label: latencies,
    'Offloading % at ' + config_label: offloading_percentage
    })
    all_results.append(data)

all_results = []
n_vehicles_options = [70]#, 20, 30, 40, 50, 60, 70, 80, 90, 100]
d_V_options = [90,50,30040, 30000,300040, 300000, 3000040, 3000000] #31kbyte, 310kbyte, and 3.1Mbyte
bar_d_V_options = [90,50,30040, 30000,300040,300000]#15.5kbyte, 155 kbyte, 1.55Mbyte 
T_upload_C_options = [100, 500, 1000] # bps and will be multiplied by 1e6 up there
T_download_C_options = [100, 500, 1000] # bps and will be multiplied by 1e6 up there (0.0003, 0.08, 0.4)
best_configs = {
    10: [(0.003, 0.01, 0.4),(0.003, 0.08, 0.4),(0.003, 0.01, 0.9),(0.003, 0.08, 0.9)],#(0.0003, 0.01, 0.4),(0.0003, 0.01, 0.9),(0.0003, 0.08, 0.9)],
    20: [(0.003, 0.01, 0.4),(0.003, 0.08, 0.4),(0.003, 0.01, 0.9),(0.003, 0.08, 0.9)],#(0.0003, 0.01, 0.4),(0.0003, 0.01, 0.9),(0.0003, 0.08, 0.9)],
    30: [(0.003, 0.01, 0.4),(0.003, 0.08, 0.4),(0.003, 0.01, 0.9),(0.003, 0.08, 0.9)],#(0.0003, 0.01, 0.4),(0.0003, 0.01, 0.9),(0.0003, 0.08, 0.9)],
    40: [(0.003, 0.01, 0.4),(0.003, 0.01, 0.9),(0.003, 0.08, 0.9)],#(0.0003, 0.01, 0.4),(0.0003, 0.01, 0.9),(0.0003, 0.08, 0.9)],
    50: [(0.003, 0.01, 0.4),(0.003, 0.08, 0.4),(0.003, 0.01, 0.9),(0.003, 0.08, 0.9)],#(0.0003, 0.01, 0.4),(0.0003, 0.08, 0.4),(0.0003, 0.01, 0.9),(0.0003, 0.08, 0.9)],
    60: [(0.003, 0.01, 0.4),(0.003, 0.08, 0.4),(0.003, 0.01, 0.9),(0.003, 0.08, 0.9)],#(0.0003, 0.01, 0.4),(0.0003, 0.08, 0.4),(0.0003, 0.01, 0.9),(0.0003, 0.08, 0.9)],
    70: [(0.003, 0.01, 0.4),(0.003, 0.08, 0.4),(0.003, 0.01, 0.9),(0.003, 0.08, 0.9)],#(0.0003, 0.01, 0.4),(0.0003, 0.08, 0.4),(0.0003, 0.01, 0.9),(0.0003, 0.08, 0.9)],
    80: [(0.003, 0.01, 0.4),(0.003, 0.08, 0.4),(0.003, 0.01, 0.9),(0.003, 0.08, 0.9)],#(0.0003, 0.01, 0.4),(0.0003, 0.08, 0.4),(0.0003, 0.01, 0.9),(0.0003, 0.08, 0.9)],
    90: [(0.003, 0.01, 0.4),(0.003, 0.08, 0.4),(0.003, 0.01, 0.9),(0.003, 0.08, 0.9)],#(0.0003, 0.01, 0.4),(0.0003, 0.08, 0.4),(0.0003, 0.01, 0.9),(0.0003, 0.08, 0.9)],
    100: [(0.003, 0.01, 0.4),(0.003, 0.08, 0.4),(0.003, 0.01, 0.9),(0.003, 0.08, 0.9)]#(0.0003, 0.08, 0.4),(0.0003, 0.01, 0.9),(0.0003, 0.08, 0.9)]
}

for n_vehicles in n_vehicles_options:
    for (d_V, bar_d_V) in zip(d_V_options, bar_d_V_options):
            for (T_upload_C, T_download_C) in zip(T_upload_C_options, T_download_C_options):
                    for config in best_configs[n_vehicles]:
                        learning_rate, entropy_coefficient, gamma = config
                        print(f'Starting experiment with {n_vehicles} vehicles, learning rate {learning_rate}, '
                              f'entropy coefficient {entropy_coefficient}, and discount factor {gamma}, '
                              f'task size {d_V}, result size {bar_d_V}, upload rate {T_upload_C}, download rate {T_download_C}')
                        run_experiment(n_vehicles, learning_rate, entropy_coefficient, gamma, [d_V], [bar_d_V], [T_upload_C], [T_download_C])

# After running the experiments and saving the combined results:
combined_results = pd.concat(all_results)
file_path = 'latency_vs_vehicles_task_result_rates_all_n=70_to_delete.csv'
combined_results.to_csv(file_path, index=False)
df_new = pd.read_csv(file_path)

# Detect the first non-NaN value in each column to determine how much to shift
shifts = df_new.apply(lambda col: col.first_valid_index())

# Shift columns up by the number of rows they are displaced
for col in df_new.columns:
    if shifts[col] is not None:
        df_new[col] = df_new[col].shift(-shifts[col])

# Drop any remaining NaN rows that resulted from the shift
df_corrected = df_new.dropna()
print('ok cleaned')

df_new = df_corrected#pd.read_csv(file_path_new)

# Detect the first non-NaN value in each column to determine how much to shift
shifts = df_new.apply(lambda col: col.first_valid_index())

# Shift columns up by the number of rows they are displaced
for col in df_new.columns:
    if shifts[col] is not None:
        df_new[col] = df_new[col].shift(-shifts[col])

# Drop any remaining NaN rows that resulted from the shift
df_corrected = df_new.dropna()

# Separate latency and offloading columns
latency_columns = [col for col in df_corrected.columns if 'Latency' in col]
offloading_columns = [col for col in df_corrected.columns if 'Offloading' in col]

# Calculate the average for latency and offloading columns
latency_averages = df_corrected[latency_columns].mean(axis=1)
offloading_averages = df_corrected[offloading_columns].mean(axis=1)

# Create a new DataFrame with the iteration and averaged latency and offloading columns
df_final = pd.DataFrame({
    'Iteration': df_corrected['Iteration']#,'Average Latency': latency_averages
})

# Add the individual latency columns
for col in latency_columns:
    df_final[col] = df_corrected[col]

# Add the individual offloading columns
for col in offloading_columns:
    df_final[col] = df_corrected[col]

# Save the final corrected and organized data
output_final_path = 'n=70_all_average.csv'
df_final.to_csv(output_final_path, index=False)

print('Data corrected, separated, and averages calculated.')
```

```python
def detect_minimizing_segments(data, window_size=30):
    """
    Detect segments where the data is minimizing over a certain window size.
    Returns a boolean mask where True indicates a minimizing segment.
    """
    decreasing_mask = (data.diff().rolling(window=window_size).sum() < 0)
    return decreasing_mask

def is_decreasing(data, window_size=30):
    """
    Check if the data is showing a decreasing trend over the entire signal.
    Returns True if the trend is generally decreasing, otherwise False.
    """
    return data.rolling(window=window_size).mean().diff().mean() < 0
```
